```python
import os
import json
import csv
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
df_process = pd.read_csv('./data/레시피_과정정보.csv', error_bad_lines=False, encoding="cp949")
df_material = pd.read_csv('./data/레시피_재료정보.csv', error_bad_lines=False, encoding="cp949")
df_food = pd.read_csv('./data/레시피_기본정보.csv', error_bad_lines=False, encoding="cp949")
    
```

```python
df_dict = {}
for index, row in df_material.iterrows() :
    if row['레시피 코드'] in df_dict:
        df_dict[row['레시피 코드']] += row['재료명'] + ','
    else :
        df_dict[row['레시피 코드']] =  row['재료명'] + ','
df_dict
```

```python
receipt_list = []
df_food['재료'] = ''

for key, value in df_dict.items():
    temp = value.split(',')
    temp = temp[:-1] #마지막 열 제거 ','
    df_dict[key] = temp 
print(df_dict)
    
for index, row in df_food.iterrows() :
    receipt_list.append(df_dict[row['레시피 코드']] )

df_food['재료'] = receipt_list
```

```python
one_hot_list = df_material['재료명'].unique()
values = []
for x in range(len(one_hot_list)):
    values.append(x)
values_len = len(values)
encoding = np.eye(values_len)[values]


temp = df_food[df_food['레시피 코드'] == 1 ]['재료']
print(temp)

one_index_list = []
change_one_list = []
for index ,row in df_food.iterrows():
    temp = row['재료']
    temp_list = []
    zero_arr = [0]* len(one_hot_list)
    for data in temp : #리스트 돌리려고 
        try :
            index = np.where(one_hot_list == data)[0][0]
            temp_list.append(index)
            zero_arr[index] = 1
        except :
            print('잘못된 재료 입력입니다 : {}'.format(data))
   
    one_index_list.append(temp_list)
    change_one_list.append(zero_arr)
df_food['재료 인덱싱'] = one_index_list
df_food['ONE HOT'] = change_one_list

```

```python
def comp_one_hot(df, code, menu_cnt) :
    comp1 = df[df['레시피 코드'] == code]['ONE HOT'].item()
    comp_dict = {}
    for index, row in df.iterrows():
        if row['레시피 코드'] == code :
            continue
        comp2 = row['ONE HOT']
        cnt = comp_pair_one_hot(comp1, comp2)
        comp_dict[row['레시피 코드']] = cnt
    print('주메뉴 :',df[df['레시피 코드']== code]['레시피 이름'].item())
    comp_list = sorted(comp_dict.items(), key=lambda x: x[1], reverse= True)
    for index in range(menu_cnt):
        print('추천 메뉴 {} :'.format(index+1) ,df[df['레시피 코드']== comp_list[index][0]] ['레시피 이름'].item())    
    
def comp_pair_one_hot(comp1, comp2) :
    cnt = 0
    for index in range(len(comp1)) :
        if comp1[index] == 1 and comp2[index] == 1 :
            cnt += 1
    return cnt

def user_list_input(user_list) :
    zero_arr = [0]* len(one_hot_list)
    for data in user_list :
        try :
            index = np.where(one_hot_list == data)[0][0]
            zero_arr[index] = 1
        except :
            print('잘못된 재료 입력입니다 : {}'.format(data))
    return zero_arr
```
